```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/data_excel/toko_online_sample.csv'
df = pd.read_csv(file_path)
display(df.head())
```

```python
df.columns.tolist()
```

```python
df.sample(5)
```

```python
df.select_dtypes(include=np.number).quantile([0.25, 0.5, 0.75])
```

```python
df.isnull().sum()
```

```python
print("print missing values per kolom")
missing = df.isnull().sum()
missing_percent = (missing/len(df))*100

missing_df = pd.DataFrame({'missing': missing, 'missing_percent': missing_percent})
print(missing_df[missing_df['missing_percent']>0].sort_values('missing_percent', ascending=False))
```

```python
missing_df[missing_df['missing_percent'] > 0].sort_values('missing_percent', ascending=False).plot(kind='bar', y='missing_percent', legend=False)
plt.title('Persentase Nilai Hilang per Kolom')
plt.xlabel('Kolom')
plt.ylabel('Persentase Nilai Hilang')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Visualisasi distribusi kolom numerik (harga, qty, diskon, rating)
df.hist(column=['harga', 'qty', 'diskon', 'rating'], bins=20, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
# Visualisasi distribusi kolom kategorikal (kota, kategori, metode_bayar, returned)
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
sns.countplot(data=df, y='kota', ax=axes[0, 0], order=df['kota'].value_counts().index)
sns.countplot(data=df, y='kategori', ax=axes[0, 1], order=df['kategori'].value_counts().index)
sns.countplot(data=df, y='metode_bayar', ax=axes[1, 0], order=df['metode_bayar'].value_counts().index)
sns.countplot(data=df, y='returned', ax=axes[1, 1], order=df['returned'].value_counts().index)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['harga'].dropna(), bins=20, edgecolor='black', range=[0, 100])
plt.title('Distribusi Harga (Range 0-100)')
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='harga', fill=True)
plt.title('Density Plot Distribusi Harga')
plt.xlabel('Harga')
plt.ylabel('Densitas')
plt.show()
```

```python
print("Transaksi dengan harga antara 20 dan 30:")
display(df[(df['harga'] >= 20) & (df['harga'] <= 30)].head())

print("\nTransaksi dengan harga antara 45 dan 55:")
display(df[(df['harga'] >= 45) & (df['harga'] <= 55)].head())
```

```python
price_counts = df['harga'].value_counts().reset_index()
price_counts.columns = ['harga', 'frekuensi']

plt.figure(figsize=(10, 6))
plt.scatter(price_counts['harga'], price_counts['frekuensi'])
plt.title('Frekuensi Setiap Harga Unik')
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```
