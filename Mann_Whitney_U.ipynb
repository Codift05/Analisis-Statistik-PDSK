```python
#  Import library yang dibutuhkan
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

#  Unggah file secara manual di Google Colab (Jalankan hanya jika file belum diunggah)
from google.colab import files
uploaded = files.upload()

#  Baca file Excel
file_path = list(uploaded.keys())[0]  # Mengambil nama file yang diunggah
df = pd.read_excel(file_path)

#  Cek nama kolom untuk memastikan tidak ada kesalahan
print("Nama Kolom:", df.columns.tolist())

#  Bersihkan nama kolom dari spasi yang tersembunyi
df.columns = df.columns.str.strip()

#  Pastikan nama kolom benar, sesuaikan dengan hasil print sebelumnya
kolom_durasi = "Jika Ya, berapa lama biasanya kamu menggunakan ponsel sebelum tidur?"
kolom_tidur = "Berapa lama rata-rata kamu tidur setiap malam?"

#  Ubah data ke tipe numerik (jika belum dalam bentuk angka)
df[kolom_durasi] = pd.to_numeric(df[kolom_durasi], errors='coerce')
df[kolom_tidur] = pd.to_numeric(df[kolom_tidur], errors='coerce')

#  Pisahkan data ke dalam dua kelompok berdasarkan durasi penggunaan ponsel
group1 = df[df[kolom_durasi] <= 30][[kolom_durasi, kolom_tidur]].dropna()
group2 = df[df[kolom_durasi] > 30][[kolom_durasi, kolom_tidur]].dropna()

#  Beri label kategori
group1.insert(0, "Kategori", "≤ 30 Menit")
group2.insert(0, "Kategori", "> 30 Menit")

#  Gabungkan dua tabel untuk tampilan lebih rapi
tabel_final = pd.concat([group1, group2], ignore_index=True)

#  Tampilkan tabel dalam format yang lebih rapi
from IPython.display import display  # Untuk tampilan lebih baik di Colab
print("\n Tabel Data Kelompok Berdasarkan Penggunaan Ponsel")
display(tabel_final)

#  Uji Mann-Whitney U
stat, p_value = stats.mannwhitneyu(group1[kolom_tidur], group2[kolom_tidur], alternative='two-sided')

#  Tampilkan hasil uji statistik
print("\n Hasil Uji Mann-Whitney U Test:")
print(f"Statistik Uji: {stat}")
print(f"P-Value: {p_value}")

#  Interpretasi hasil
alpha = 0.05
if p_value < alpha:
    print("\n Hasil: Terdapat perbedaan signifikan antara dua kelompok.")
else:
    print("\n Hasil: Tidak terdapat perbedaan signifikan antara dua kelompok.")

#  Visualisasi Data

#  Boxplot untuk melihat distribusi tidur di kedua kelompok
plt.figure(figsize=(8, 5))
sns.boxplot(x=tabel_final["Kategori"], y=tabel_final[kolom_tidur], palette=["blue", "red"])
plt.xlabel("Kategori Penggunaan Ponsel")
plt.ylabel("Durasi Tidur (Jam)")
plt.title("Distribusi Durasi Tidur Berdasarkan Penggunaan Ponsel")
plt.grid(True)
plt.show()

#  Histogram untuk melihat sebaran data
plt.figure(figsize=(8, 5))
sns.histplot(group1[kolom_tidur], color="blue", label="≤ 30 Menit", kde=True, bins=10)
sns.histplot(group2[kolom_tidur], color="red", label="> 30 Menit", kde=True, bins=10)
plt.xlabel("Durasi Tidur (Jam)")
plt.ylabel("Frekuensi")
plt.title("Histogram Durasi Tidur")
plt.legend()
plt.grid(True)
plt.show()

#  Bar Chart untuk melihat rata-rata durasi tidur setiap kelompok
plt.figure(figsize=(6, 4))
avg_sleep = [group1[kolom_tidur].mean(), group2[kolom_tidur].mean()]
labels = ["≤ 30 Menit", "> 30 Menit"]
plt.bar(labels, avg_sleep, color=["blue", "red"])
plt.ylabel("Rata-rata Durasi Tidur (Jam)")
plt.title("Perbandingan Rata-rata Durasi Tidur")
plt.grid(axis="y")
plt.show()
```
